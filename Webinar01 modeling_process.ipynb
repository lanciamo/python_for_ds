## Общее описание процесса построения прогнозных моделей

* Постановка задачи (business understanding, data understanding)
    - описание проблемы
    - формализация бизнес-требований
    - оценка имеющихся данных и ресурсов
    - формирование гипотез
    - выбор метрики качества и критериев приёмки
    - определение сроков


* Сбор и загрузка данных (data collection)
    - парсинг
    - базы данных
    - файловые хранилища


* Подготовка и разметка обучающей выборки (train dataset preparation)


* Обзор обучающего датасета (train dataset overview)
    - проверка, что данные загрузились верно (кодировка, структура  и т.д.)
    - оценка типа данных
    - оценка объемов данных
    
    
* Предобработка (data prepararion)
    - приведение типов
    - обработка пропусков
    - обработка выбросов
    - балансировка обучающих примеров
    
    
* Работа с признаковым пространством (feature selection, extraction and engineering)
    - отбор признаков
    - построение новых признаков
    - снижение / повышение размерности
        
    
* Анализ (data analysis / data exploration)
    - визуализация данных
    - поиск зависимостей в признаках
    - поиск инсайтов
    - проверка гипотез
    
    
* Моделирование (modeling)
    - выбор подходящих моделей и функции ошибки
    - предварительное моделирование и отбор более успешных моделей
    - получение бейзлайна
    - переобучение и оценка обобщающей способоности модели
    - выбор лучшей модели и настройка гиперпараметров
    
    
* Оценка результатов (evaluation)
    - оценка качества прогнозирования
    - оценка прочих критериев приёмки

 
* Внедрение полученного решения (implementation / deployment)
