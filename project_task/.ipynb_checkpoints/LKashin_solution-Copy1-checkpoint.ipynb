```python
plt.figure(figsize = (16, 4))

plt.subplot(121)
df_train['HouseYear'].hist()
plt.ylabel('count')
plt.xlabel('HouseYear')

plt.subplot(122)
sns.kdeplot(df_train['HouseYear'], shade=True, legend=False)
plt.xlabel('HouseYear')

plt.suptitle('Distribution of HouseYear')

plt.show()
```
